importing the csv file of the sample problem

```python
import pandas as pd
from IPython.display import display

df = pd.read_csv('SAMPLE PROBLEM.csv')
df = df.reset_index(drop=True)
display(df)
```

GRAPH

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.scatter(df['x'], df['y'], color='blue', label='Data Points')
m, b = np.polyfit(df['x'], df['y'], 1)
plt.plot(df['x'], m * df['x'] + b, color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regression Line')
plt.legend()
plt.show()
```

INTERPRETATION:

The negative slope of the regression line signifies a negative correlation between x and y, demonstrating an inverse relationship where y decreases as x increases.

NOTE: An inverse relationship on a graph is shown by a negative slope on a linear graph or downward trending curve. An inverse relationship occurs when two variables change in opposite directions. For example, when X increases, Y decreases. according to GOOGLE
