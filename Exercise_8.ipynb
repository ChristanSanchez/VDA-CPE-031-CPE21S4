#Plotting Without a Line

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0,6])
ypoints = np.array([0,250])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

#Multiple Points

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

#Default

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker='x')
plt.show()
```

Format Strings fmt

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints,'o:r') #marker | line | color (RGB)
plt.show()
```

#Marker Size

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints,'o:r', ms=20) #marker | line | color (RGB)
plt.show()
```

#Marker Color

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints,'o:r', ms=20,mec='b') #marker | line | color (RGB)
plt.show()
```

Marker Face Color or mfc

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints,'o:r', ms=20,mec='b',mfc='g') #marker | line | color (RGB)
plt.show()
```

#Dash Line

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints,linestyle='dashed') #marker | line | color (RGB)
plt.show()
```

#Matplot labes and Titles

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.title("Distance Travelled Everyday", color = 'r')
plt.xlabel("Distance in km")
plt.ylabel("Speed in km/s")
plt.show()
```

#Matplot grid - you can use grid() to add gridlines

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.title("Distance Travelled Everyday", color = 'r')
plt.xlabel("Distance in km")
plt.ylabel("Speed in km/s")
plt.grid(axis='y')
plt.show()
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.title("Distance Travelled Everyday", color = 'r')
plt.xlabel("Distance in km")
plt.ylabel("Speed in km/s")
plt.grid(axis='x',linestyle='--',linewidth=0.50)
plt.show()
```

#Show Multiple Plots

```python
#plot 1
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1)
plt.plot(xpoints, ypoints)

#2
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([10, 20, 30, 40])
plt.subplot(1,2,2)
plt.plot(xpoints,ypoints)

plt.grid(axis = 'x', linestyle = '--',linewidth=0.50)
plt.show()
```

#Matplot Scatter

```python

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.scatter(xpoints, ypoints)


xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([10, 20, 30, 40])

plt.scatter(xpoints,ypoints)

plt.show()
```

#Matplot Bar
- you can use the bar() to draw bars

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['Student1', 'Student2', 'Student3'])
y = np.array([90,85,86])
plt.bar(x,y)
plt.show
```

#Matplotlib Pie Chart - You can use pie() to draw pie chart

```python
g = np.array([35, 25, 25, 15])
grader = [ "Grade 1 = 35%","Grade 2 = 25%","Grade 3 = 25%","Grade 4 = 15%"]
plt.pie(g,labels=grader)
plt.legend(title = 'Grader',labels = grader)
plt.show()
```
