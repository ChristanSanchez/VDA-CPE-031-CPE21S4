a) Read each device's data of release and sales. Show it in a line plot

```python
import matplotlib.pyplot as plt

df = pd.read_csv('Technological-Products-Sample-Data - Tech Products.csv')
plt.figure(figsize=(15, 7))

plt.title("Sales by Device Over Years")
plt.xlabel("Date of Release")
plt.ylabel("Sales (USD)")
plt.plot(df['Date of Release'], df['Price (USD)'])
plt.legend(loc='lower right')
plt.xticks(rotation=45)
plt.show()
```

b) Get the total sales and show line plot with the following style properties.

- Line style dotted and Line-color should be red
- Show legend at the lower right location.
- X label name = Year
- Y label name = Sales
- Add a circle marker
- Line marker color as red
- Line width should be 3

```python
import matplotlib.pyplot as plt


total_sales = df['Price (USD)'].sum()
print(f"Total Sales: {total_sales}")

# Plotting the data
plt.figure(figsize=(15, 7))
plt.plot(df['Date of Release'], df['Price (USD)'], linestyle=':', color='red', linewidth=3, marker='o', markerfacecolor='red', markersize=8, label='Sales Data')
plt.title("Sales by Device Over Years", color='r')
plt.xlabel("Year")
plt.ylabel("Sales")
plt.legend(loc='lower right')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

c) Read all product sales data and show it using a multiline plot. Label Them

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Technological-Products-Sample-Data - Tech Products.csv')

df['Date of Release'] = pd.to_datetime(df['Date of Release'])
df['Release Month'] = df['Date of Release'].dt.to_period('M')

plt.figure(figsize=(15, 5))
plt.title('Sales Data for All Products')
plt.xlabel('Release Month')
plt.ylabel('Total Sales (USD)')

for product in df['Model'].unique():
    product_data = df[df['Model'] == product].groupby('Release Month')['Price (USD)'].sum().reset_index()
    plt.plot(product_data['Release Month'].astype(str), product_data['Price (USD)'], label=product)

plt.xticks(rotation=45, ha='right')
plt.legend(loc='best', fontsize='small')
plt.tight_layout()
plt.show()
```

d) Read each brand sales data and show it using the bar chart. Label Them

```python
# Group by brand and sum the sales
brand_sales = df.groupby('Brand')['Price (USD)'].sum()

# Plotting the data
plt.figure(figsize=(15, 7))
brand_sales.plot(kind='bar', color='skyblue')
plt.title("Total Sales by Brand")
plt.xlabel("Brand")
plt.ylabel("Total Sales (USD)")
plt.xticks(rotation=45)
plt.show()
```

e) read each device and show it using the histogram to see the most common Country of Origin. Label them.

```python
# Plotting the data
plt.figure(figsize=(15, 7))
df['Country of Origin'].value_counts().plot(kind='bar', color='lightgreen')
plt.title("Most Common Country of Origin")
plt.xlabel("Country of Origin")
plt.ylabel("Number of Devices")
plt.xticks(rotation=45)
plt.show()
```

f) Create a pie chart that shows each brand Country of Origin. Label them

```python
# Group by country of origin and count the brands
country_origin = df['Country of Origin'].value_counts()

# Plotting the data
plt.figure(figsize=(10, 10))
country_origin.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title("Country of Origin Distribution")
plt.ylabel('')
plt.show()
```
