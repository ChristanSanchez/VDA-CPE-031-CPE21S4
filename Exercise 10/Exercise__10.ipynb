# IMPORTING THE CSV FILE

```python
import pandas as pd
from IPython.display import display

df = pd.read_csv('Proficiency_Exam_vs_Course_Grade.csv')
df = df.reset_index(drop=True)
display(df)
```

PLOTTING OF THE DATA GIVEN

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


plt.scatter(df['Proficiency Exam Score'], df['Course Grade'], color='blue', label='Data Points')
m, b = np.polyfit(df['Proficiency Exam Score'], df['Course Grade'], 1)
plt.plot(df['Proficiency Exam Score'], m * df['Proficiency Exam Score'] + b, color='red')
plt.xlabel('Proficiency Exam Score (X)')
plt.ylabel('Course Grade (Y)')
plt.title('Proficiency Exam Score Vs Course Grade')
plt.legend()
plt.show()
```

INTERPRETATION:

The graph shows that students who tend to score higher in the Exam have higher grades in the course code
